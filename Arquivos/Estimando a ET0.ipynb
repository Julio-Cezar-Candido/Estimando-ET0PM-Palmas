```python
import pandas as pd
import numpy as np
from tqdm import trange
from time import sleep
```

```python
#Abrindo Arquivo
palmas = pd.read_excel("Palmas_2021.xlsx")
palmas
```

```python
datas = []
prec = []
pres = []
radi = []
temp_maxi = []
temp_mini = []
umid = []
umid_maxi = []
umid_mini = []
velo_vent = []
eto_pm = []
```

```python
#Pegando dados da Tabela e Trattando:
cont_linh_tabe = -1

#Este laço entra em cada mês da tabela palmas:
for mes in trange(1, 9):
    sleep(0.1)
    
    #Define quantidade de 31 dias para os meses [1, 3, 5, 7, 8, 10, 12]:
    if mes in [1, 3, 5, 7, 8, 10, 12]: 
        dia_do_mes = 32
        
    #Define quantidade de 28 dias para o mes 2:
    elif mes == 2: 
        dia_do_mes = 29
    
    #Define quantidade de 30 dias para os meses [4, 6, 9, 11]:
    else: 
        dia_do_mes = 31
    
    #Este laço entra em cada dia do mes:
    for dia in range(1, dia_do_mes):
        
        #Zerando auxiliares:
        prec_aux = 0
        pres_aux = []
        radi_aux = 0
        umid_aux = []
        umid_maxi_aux = []
        umid_mini_aux = []
        velo_vent_aux = []
        
        #Este laço entra em cada hora do dia:
        for hora in range(24):
            
            #colocando contador para cada linha:
            cont_linh_tabe += 1
            
           #Faz o somatorio da precipitação diaria (mm) de cada hora com dados:
            if palmas["PRECIPITAÇÃO (mm)"][cont_linh_tabe] != "Sem dados":
                prec_aux += palmas["PRECIPITAÇÃO (mm)"][cont_linh_tabe]
                
            #Salva as pressões horarias em uma lista:
            if palmas["PRESSAO ATMOSFERICA"][cont_linh_tabe] != "Sem dados": 
                pres_aux.append(palmas["PRESSAO ATMOSFERICA"][cont_linh_tabe])
                
            #Faz o somatorio da radiação global em (Mj/m²) de cada hora com dados:
            if palmas["RADIACAO GLOBAL (Kj/m²)"][cont_linh_tabe] != "Sem dados": 
                radi_aux += ( palmas["RADIACAO GLOBAL (Kj/m²)"][cont_linh_tabe] * 0.001)
            
            #Coleta a Tmax de cada dia:
            if hora == 18:
                temp_maxi.append(palmas["TEMPERATURA MÁXIMA (°C)"][cont_linh_tabe])
                
            #Coleta a Tmin de cada dia:
            if hora == 9: 
                temp_mini.append(palmas["TEMPERATURA MÍNIMA (°C)"][cont_linh_tabe])
                
            #Salva as umidades horarias em uma lista:
            if palmas["UMIDADE (%)"][cont_linh_tabe] != "Sem dados": 
                umid_aux.append(palmas["UMIDADE (%)"][cont_linh_tabe])
                
            #Salva as umidades maximas horarias em uma lista:
            if palmas["UMIDADE MAX. (%)"][cont_linh_tabe] != "Sem dados": 
                umid_maxi_aux.append(palmas["UMIDADE MAX. (%)"][cont_linh_tabe])
                
            #Salva as umidades minimas horarias em uma lista:
            if palmas["UMIDADE MIN. (%)"][cont_linh_tabe] != "Sem dados": 
                umid_mini_aux.append(palmas["UMIDADE MIN. (%)"][cont_linh_tabe])
            
            #Salva as velocidades do vento horarias em uma lista (m/s):
            if palmas["VENTO (m/s)"][cont_linh_tabe] != "Sem dados": 
                velo_vent_aux.append(palmas["VENTO (m/s)"][cont_linh_tabe])
                
            #pegando as datas:
            if hora == 0:
                datas.append(palmas["Data"][cont_linh_tabe])
                
        #Salva o somatorio da precipitação de cada dia em uma lista em (mm/dia):
        prec.append(prec_aux)
        
        #Salva a média da Pressão de cada dia em uma lista:
        pres.append(np.mean(pres_aux))
        
        #Salva o somatorio da radiação global de cada dia em uma lista em (Mj/m²):
        radi.append(radi_aux)
        
        #Salva a média da Umidade de cada dia em uma lista em (%):
        umid.append(np.mean(umid_aux))
        
        #Salva a média da Umidade maxima de cada dia em uma lista em (%):
        umid_maxi.append(np.mean(umid_maxi_aux))
        
        #Salva a média da Umidade minima de cada dia em uma lista em (%):
        umid_mini.append(np.mean(umid_mini_aux))
        
        #Salva a média da Velocidade do vento de cada dia em uma lista em (m/s):
        velo_vent.append(np.mean(velo_vent_aux)) 
```

```python
#Pegando a latitude e convertendo em Radianos:
latitude = palmas["LATITUDE"][0]
latitude = np.radians(latitude)
```

```python
#Calculando a ETO de cada dia:

#Entrando em cada dia e estimando a et0:
for dia in trange(243):
    sleep(0.01)
    
    if temp_maxi[dia] == "Sem dados" or temp_mini[dia] == "Sem dados":
        et0PM = "Sem dados"
    else:
        e0max = (6.1078 * (10.0 ** ( (7.5 * temp_maxi[dia]) / (237.3 + temp_maxi[dia]) ) ) ) / 10.0

        e0min = (6.1078 * (10.0 ** ( (7.5 * temp_mini[dia]) / (237.3 + temp_mini[dia]) ) ) ) / 10.0

        es = (e0min + e0max) / 2.0

        ea = ( (e0min * (umid_maxi[dia] / 100.0) ) + (e0max * (umid_mini[dia] / 100.0) ) ) / 2.0

        u2 = velo_vent[dia] * ( (np.log( (2.0 - 0.08) / (0.01476) ) ) / np.log( (10.0 - 0.08) / (0.01476)) )

        delta = ( (4098.0 * es) / (  ( ( (temp_maxi[dia] + temp_mini[dia]) / 2.0) + 237.3) ** 2.0) )

        #teste_pressao = ( 1013.25 - (0.106 * altitude) ) / 10 #verificar

        gama = ( (0.0016286 * prec[dia]) / 2.45)

        inclinacao = np.radians(23.45) * np.sin (  ( (np.radians(360.0) / 365.0) * (284.0 + (dia + 1) )  ) )

        hn = np.arccos(  ( -np.tan(latitude) ) * np.tan(inclinacao) )

        radAtm = 37.6 * ( ( hn *  np.sin(latitude) * np.sin(inclinacao) ) + ( np.cos(latitude) * np.cos(inclinacao) * np.sin(hn) )  )

        rs = (0.75 + ( 2.0 * (10.0  ** -5) ) * 280.0 ) * radAtm

        rol = 4.9 * (10.0 ** -9.0) * ( ( ( ( (temp_maxi[dia] + 273.0)  ** 4.0) ) + ( (temp_mini[dia] + 273.0) ** 4.0)  ) / 2.0 ) * (  0.34 - (0.14 * (np.sqrt(ea) ) ) )  * (1.35 * (radi[dia] / rs) - 0.35) 

        rn = radi[dia] -(0.23  * radi[dia]) - rol

        et0PM = (((0.408  *  delta * rn) + gama * (900.0 / ( ( (temp_maxi[dia] + temp_mini[dia]) / 2.0) + 273.0)) * u2  * (es - ea)) / (delta + (gama * (1.0 + 0.34 * u2) ) ) )

    eto_pm.append(et0PM)
```

```python
#Salvar dados em uma tabela
#Criando um dict para organizar o DataFrame
vari_palm = {
    'DATA' : datas,
    'PRECIPITAÇÃO (mm/dia)' : prec,
    'PRESSÃO ATMOSFERICA' : pres,
    'RADIACAO GLOBAL (Mj/m²/dia)' : radi,
    'TEMPERATURA MÁXIMA (°C)' : temp_maxi,
    'TEMPERATURA MÍNIMA (°C)' : temp_mini,
    'UMIDADE (%)' : umid,
    'UMIDADE MAX. (%)' : umid_maxi,
    'UMIDADE MIN. (%)' : umid_mini,
    'VELOCIDADE DO VENTO 10m (m/s)' : velo_vent,
    'ETO_PMFAO (mm/dia)' : eto_pm,
}

#Criando DataFrame:
df = pd.DataFrame(data=vari_palm)

#Salvando arquivo Excel
df.to_excel("ETOPM_PALMAS_2021.xls", index=False)
```
